```python
1. What is a random variable in probability theory?
-> In probability theory, a random variable is a function that assigns a numeric
al value to each outcome of a random experiment.

Example-

Experiment: Toss a fair coin twice.

2. What are the types of random variables?

-> The types of random variables are:

A.) Discrete Random Variable

Takes countable values (finite or countably infinite).

Often results from counting something.

Probability is described using a Probability Mass Function (PMF).

Example:
Number of heads in 5 coin tosses (0,1,2,3,4,5)
Number of customers in a shop at a given hour

B.) Continuous Random Variable

Takes values from an uncountable set (usually an interval of real numbers).

Often results from measuring something.

Probability is described using a Probability Density Function (PDF).

Example:
Height of a person (e.g., 170.2 cm, 170.21 cm, …)
Time taken to run a race (e.g., 12.43 seconds)

C.) Mixed Random Variable

Has both discrete and continuous parts.

Example:
Time until the next bus arrives:
Could be exactly 0 minutes if a bus is already at the stop (discrete point).

Otherwise, a continuous range of waiting times.

3. Explain the difference between discrete and continuous distributions

1. Discrete distributions

-Think of counting things.
-You can list the possible values one by one.
-There are "gaps" between the possible values — no in-betweens.

Example: The number of cars in a parking lot could be 0, 1, 2, 3… but not 2.5
cars.

Analogy: Counting coins in your pocket — you can have 4 coins or 5 coins, but
not 4.3 coins.

2. Continuous distributions

-Think of measuring things.
-Values can be any number in a range, including fractions and decimals.
-No gaps — between any two numbers, there are infinitely many more numbers.

Example: Your height could be 170 cm, 170.1 cm, 170.12 cm, etc.

Analogy: Measuring water in a glass — you can have 200 ml, 200.1 ml, 200.01
ml, there's no limit to how precise you can get.

The key difference:

-Discrete = Countable, whole steps.
-Continuous = Measurable, smooth range.

4. What is a binomial distribution, and how is it used in probability?

-> A binomial distribution is a probability distribution that tells you the like
lihood of getting a certain number of successes in a fixed number of independent
trials, when each trial has only two possible outcomes (success or failure) and
the probability of success stays the same.

When to use it:

You use a binomial distribution when:

-Fixed number of trials (n) — e.g., flipping a coin 10 times.

-Two outcomes per trial — success or failure (heads/tails, win/lose, pass/fail).

-Independent trials — outcome of one trial doesn't affect others.

-Constant probability of success (p) for each trial.

5. What is the standard normal distribution, and why is it important?

-> The standard normal distribution is a special case of the normal
   (bell-shaped) distribution that is:

-Perfectly symmetric around the mean
-Has mean = 0
-Has standard deviation = 1
-Mathematically, it is denoted as N(0,1).

Why is it important?

a) It is the reference model for many statistical methods
-Many real-world measurements (like heights, exam scores, errors in measurement)
tend to follow a bell-shaped pattern.
-If we convert them into the standard normal form, we can compare and analyze
them more easily.

b) Z-scores make different datasets comparable
-Any normal distribution can be transformed into a standard normal distribution

c) Probability lookup using Z-tables
Once standardized, you can use a Z-table to find probabilities or percentiles
without recalculating each time.

6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

The Central Limit Theorem (CLT) is one of the most important ideas in statistics
because it explains why normal distributions show up everywhere and why we can
use them for so many problems.

> Why it is critical ?

a) Makes statistical inference possible

-Most real-world populations are not perfectly normal.

-CLT allows us to use normal distribution methods (Z-scores, confidence intervals
, hypothesis testing) for sample means, even if the population is not normal.

b) Predictable behavior of averages

-Even if individual values are very random, their averages behave in a nice,
predictable way.

c) Foundation for many statistical tools

- t-tests, ANOVA, regression, and many other techniques rely on the CLT in the
background.

Example

-Let’s say daily sales in a shop are not normal — they’re skewed because some
rare days have extremely high sales.

If you take one day's sales, the data looks messy.

But if you take the average sales over 50 days and repeat that process many
times, those averages will form a bell curve.

This means you can safely apply normal distribution formulas to those averages.

7. What is the significance of confidence intervals in statistical analysis?

-> A confidence interval (CI) is like a range of plausible guesses for a
population parameter (such as a mean or proportion) based on sample data — but
with a stated level of certainty.

> Why they’re significant

-They show both the estimate and the uncertainty

A single number (point estimate) hides how uncertain we are.

A CI shows the range — narrow CI = more precise estimate, wide CI = less
precision.

-They help in decision-making

If the CI for a mean difference doesn’t include 0, it suggests a statistically
significant difference.

In business, medicine, and policy-making, CIs help determine if effects are
large enough to matter.

-They connect sample data to the population

They allow us to infer about the whole population without measuring everyone.

-Better than just p-values

A p-value says whether an effect exists; a CI shows how big the effect might be.


Example:
Medical trial: New drug reduces blood pressure by 5 mmHg on average, 95% CI:
(3 mmHg, 7 mmHg).

We are 95% sure the true average drop is between 3 and 7 mmHg.

Since the entire interval is above 0, it is likely the drug has a real effect.

8. What is the concept of expected value in a probability distribution?

The expected value (also called the mean of a probability distribution) is the
long-term average you’d expect if you repeated a random experiment many times.

It’s like the center of gravity of a probability distribution.

Example (Discrete)

You roll a fair six-sided die.
Possible outcomes: 1,2,3,4,5,6
Each has probability - 1/6.

-> Why it’s important ?

-Decision-making under uncertainty

Businesses use it to decide which investment or project yields the best average
return.

-Gambling & risk analysis

Expected value shows if a game is favorable or not.

-Statistics & economics

Used in insurance, stock market models, quality control, and more.

```

```python
#9.

import numpy as np
import matplotlib.pyplot as plt

# Parameters
mean = 50
std_dev = 5
n_samples = 1000

# Generate random numbers from a normal distribution
data = np.random.normal(mean, std_dev, n_samples)

# Compute mean and standard deviation
computed_mean = np.mean(data)
computed_std_dev = np.std(data)

# Print results
print(f"Computed Mean: {computed_mean:.2f}")
print(f"Computed Standard Deviation: {computed_std_dev:.2f}")

# Plot histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black', density=True)
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#10.

import numpy as np
from scipy import stats

# Given sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to NumPy array
sales_array = np.array(daily_sales)

# Calculate sample mean and standard deviation
mean_sales = np.mean(sales_array)
std_dev_sales = np.std(sales_array, ddof=1)  # ddof=1 for sample std dev
n = len(sales_array)

# Standard error
standard_error = std_dev_sales / np.sqrt(n)

# Confidence level
confidence_level = 0.95
z_score = stats.norm.ppf((1 + confidence_level) / 2)  # 1.96 for 95%

# Confidence interval
ci_lower = mean_sales - z_score * standard_error
ci_upper = mean_sales + z_score * standard_error

# Print results
print(f"Sample Mean Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```
