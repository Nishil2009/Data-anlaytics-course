1. What is hypothesis testing in statistics?

-> Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It involves formulating two competing statements (hypotheses) and using sample data to determine which statement is more consistent with the evidence.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

-Null Hypothesis (H₀): A statement that assumes no effect, no difference, or the status quo. It is what we try to disprove.

-Alternative Hypothesis (H₁ or Ha): A statement that contradicts the null hypothesis, representing an effect, difference, or change.

Example:

H₀: The average height of students = 170 cm

H₁: The average height of students ≠ 170 cm

3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

-> Significance level (α): The probability of rejecting the null hypothesis when it is actually true.

Common values: 0.05, 0.01.

Role: If the p-value ≤ α, we reject H₀ (evidence supports H₁). If p-value > α, we fail to reject H₀ (not enough evidence).

4. What are Type I and Type II errors? Give examples of each.

Type I Error (False Positive): Rejecting H₀ when it is true.

Example: A medical test says a healthy patient has a disease.

Type II Error (False Negative): Failing to reject H₀ when it is false.

Example: A medical test fails to detect a disease in a sick patient.

5. What is the difference between a Z-test and a T-test? Explain when to use each.

Z-test:

-Used when population variance (σ²) is known.

-Large sample size (n > 30).

-Assumes data is normally distributed.

T-test:

-Used when population variance is unknown.

-Suitable for small sample sizes (n < 30).

-Uses sample standard deviation instead of population standard deviation.

```python
#7
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized mean
mu_0 = 50

# Calculate sample mean and std
mean = np.mean(sample_data)
std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z statistic
z = (mean - mu_0) / (std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Sample Mean:", mean)
print("Z-statistic:", z)
print("p-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject H0: The mean is significantly different from 50.")
else:
    print("Fail to reject H0: No significant evidence that mean differs from 50.")
```

```python
#8
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate normal data
np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=200)

# Mean and std
mean = np.mean(data)
sem = stats.sem(data)

# 95% confidence interval
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

print("Mean:", mean)
print("95% Confidence Interval:", ci)

# Plot histogram
plt.hist(data, bins=20, alpha=0.7, edgecolor="black")
plt.axvline(mean, color="red", linestyle="dashed", label="Mean")
plt.title("Normal Distribution with 95% CI")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()



#9 Z-Scores and Histogram

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

# Example dataset
data = np.random.normal(50, 10, 100)
z_scores = calculate_z_scores(data)

# Plot histogram of standardized data
plt.hist(z_scores, bins=20, edgecolor="black", alpha=0.7)
plt.title("Z-Scores Histogram")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.show()
```
